# Spotify Data Exploration

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('tracks.csv')
df_tracks.head()
```

```python
df_tracks.info()
```

```python
df_tracks.describe().transpose()
```

```python
pd.isnull(df_tracks).sum()
```

### 20 Least Popular Songs in the Spotify Dataset

```python
sorted_df = df_tracks.sort_values('popularity', ascending = True).head(20)
sorted_df
```

### Top Ten Popular Songs With Popularity More Than 90.

```python
most_popular = df_tracks.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

## Top 50 Loudest Tracks

```python
top_five_loudest_tracks = df_tracks[["loudness", "name"]].sort_values(by="loudness", ascending=True)[:50]
top_five_loudest_tracks
```

```python
plt.figure(figsize=(20,15))
sns.barplot(x="loudness", y="name", data=top_five_loudest_tracks)
plt.title('Top 5 Loudest Tracks')
```

### Artists with 10 most danceable songs 

```python
most_danceable_songs=df_tracks[["danceability", "name", "artists"]].sort_values(by="danceability", ascending=False)[:10]
most_danceable_songs
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="danceability", y="artists", data=most_danceable_songs)
plt.title('Artists with 10 most danceable songs')
```

```python
df_tracks.set_index('release_date', inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
```

```python
df_tracks.head()
```

```python
df_tracks[["artists"]].iloc[18]
```

```python
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
df_tracks.duration.head()
```

### Correlation Map

```python
corr_df=df_tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",center=0,cmap="Blues",linewidths=1,linecolor="Black")
heatmap.set_title("Correlation Heatmap between variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

### Regression Plot Between Loudness and Energy

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df,y="loudness",x="energy",color="c").set(title="Loudness vs Energy Correlation")
```

### Regression Plot Between Popularity and Acousticness

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df,y="popularity",x="acousticness",color="b").set(title="Popularity vs Acousticness Correlation")
```

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="Number of songs per year")
```

```python
total_dr = df_tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title="Year vs Duration")
plt.xticks(rotation=90)
```

###  Line Graph to Show the Duration of the Songs for Each Year.

```python
total_dr = df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims = (10,5)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x = years, y = total_dr, ax = ax).set(title="Year vs Duration")
plt.xticks(rotation=90)
```

# Data Analysis Based on Genres of the Songs

```python
df_genre = pd.read_csv("SpotifyFeatures.csv")
```

```python
df_genre.head()
```

```python
df_genre.info()
```

```python
df_genre.describe().transpose()
```

```python
pd.isnull(df_genre).sum()
```

### Duration of Songs in Different Genres

```python
plt.title("Duration of the songs in Different Genres")
sns.color_palette("rocket", as_cmap=True)
sns.barplot(y='genre', x='duration_ms', data=df_genre)
plt.xlabel('Duration in milli seconds')
plt.ylabel('Genres')
```

### Top Five genres by Popularity

```python
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
famous = df_genre.sort_values("popularity", ascending=False).head(20)
sns.barplot(y='genre', x='popularity', data = famous).set(title="Top 5 Genres by Popularity")
```

### 10 most instrumental songs

```python
most_instrumental_songs=df_genre[["instrumentalness", "artist_name", "track_name"]].sort_values(by="instrumentalness", ascending=False)[:20]
most_instrumental_songs
```

```python
plt.figure(figsize=(12,7))
plt.pie(x="instrumentalness", data=most_instrumental_songs,autopct='%1.1f%%',labels=most_instrumental_songs.track_name)
plt.title('Artists with 10 most instrumental songs')
```
